```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Example data points
x = np.array([1, 2, 3])
y = np.array([2, 3, 5])

# Compute the best fitting line using linear regression (least squares)
coefficients = np.polyfit(x, y, 1)
a = coefficients[0]
b = coefficients[1]

# Generate points for the fitted line
x_fit = np.linspace(0, 4, 100)
y_fit = a * x_fit + b

# Plot the data points and the fitted line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', label='Data Points')
plt.plot(x_fit, y_fit, label=f'Fitted Line: y = {a:.2f}x + {b:.2f}', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Best Fitting Line for Example Data Points')
plt.legend()
plt.grid(True)

# Save the plot
plt.show()
```
